```python
import cv2
import numpy
import numpy as np


image = cv2.imread('small2.jpg')


if image is None:
    print("Error: Could not open or find the image.")
else:
    #  dimensions of the image
    height, width, channels = image.shape

    print(f"Image Width: {width} pixels")
    print(f"Image Height: {height} pixels")
    print(f"Number of Channels (e.g., 3 for RGB): {channels}")



image_sized = cv2.resize(image,(500,500)) #image  # cv2.resize(image,(768,500))
image = image_sized

if image is None:
    print("Error: Could not open or find the image.")
else:
    #  dimensions of the image
    height, width, channels = image.shape

    print(f"Image Width: {width} pixels")
    print(f"Image Height: {height} pixels")
    print(f"Number of Channels (e.g., 3 for RGB): {channels}")
# Define the rotation angle in degrees
#angle_degrees = 5

# angle from degrees to radians
#x  = np.deg2rad(angle_degrees)


vw = width
vh = height
height,width = image.shape[:2]
source = cv2.VideoWriter_fourcc(*'XVID')

#duration_factor = 2  # Increase the duration by a factor of 2
#frame_rate = 30.0 / (10 * duration_factor)
#frame_rate = 0.30  # Set the desired frame rate
desired_duration =3
# Calculate the number of frames needed for a 20-second duration
num_frames = 50
frame_rate = num_frames / desired_duration
rotation_video = cv2.VideoWriter('rotation_video.avi', source, frame_rate, (vw,vh)) 
scalling_video = cv2.VideoWriter('scalling_video.avi', source, frame_rate, (vw,vh))
translation_video = cv2.VideoWriter('translation_video.avi', source, frame_rate, (vw,vh))
vertical_shear_video = cv2.VideoWriter('vertical_shear_video.avi', source, frame_rate, (vw,vh))
Horizontal_shear_video = cv2.VideoWriter(' Horizontal_shear_video.avi', source, frame_rate, (vw,vh))
zoom_video = cv2.VideoWriter('zoom_video.avi', source, frame_rate, (vw,vh))
Shrinking_video = cv2.VideoWriter('Shrinking_video.avi', source, frame_rate, (vw,vh))



for a in range(0, 720,1):  # Rotation
    angle_degrees = a #* 360 / num_frames
    matrix_rotation  = cv2.getRotationMatrix2D((height/2,width/2),angle_degrees,0.5)
    rotation = cv2.warpAffine(image,matrix_rotation,(image.shape[1],image.shape[0]))
    rotation_video.write(rotation)

for s in np.arange(1, 300, 0.01): # scalling
    matrix_scalling = numpy.float32([[s,0,0],[0,s,0]])
    Scalling = cv2.warpAffine(image,matrix_scalling,(image.shape[1],image.shape[0]))
    scalling_video.write(Scalling)
    
for t in np.arange (0,300,1):#matrix_translation 300,1
    matrix_translation = numpy.float32([[1,0,t],[0,1,t]])
    translation = cv2.warpAffine(image,matrix_translation,(image.shape[1],image.shape[0]))
    translation_video.write(translation)
    
for v in np.arange (0.0,100,0.01):#shearing vertical
    matrix_vertical_shear = numpy.float32([[1,v,0],[0,1,0]])
    vertical_shear= cv2.warpAffine(image,matrix_vertical_shear,(image.shape[1],image.shape[0]))
    vertical_shear_video.write(vertical_shear)
    
for h in np.arange (0.0,10.1,0.1):#shearing Horiz
    matrix_Horizontal_shear = numpy.float32([[1,0,0],[h,1,0]])
    Horizontal_shear= cv2.warpAffine(image,matrix_Horizontal_shear,(image.shape[1],image.shape[0]))
    Horizontal_shear_video.write(Horizontal_shear)
    

#matrix_Identity = numpy.float32([[1,0,0],[0,1,0]])
#matrix_scalling = numpy.float32([[2,0,0],[0,2,0]])
#matrix_rotation = numpy.float32([[np.cos(x),-np.sin(x),0],[np.sin(x),np.cos(x),0]])
#matrix_translation = numpy.float32([[1,0,100],[0,1,100]])

for z in np.arange (0.0,10,0.01): #zooming
   
    zoom_matrix = np.float32([[z, 0, 0],[0, z, 0]])
    zoomed_image = cv2.warpAffine(image, zoom_matrix, (image.shape[1], image.shape[0]), flags=cv2.INTER_NEAREST)


   
    
    zoom_video.write(zoomed_image)





for k in np.arange (5, 0, -0.01):# Shrinking (Scaling Down)
   
    shrink_matrix = np.float32([[k, 0, 0],[0, k, 0]])
    shrunk_image = cv2.warpAffine(image, shrink_matrix, (image.shape[1], image.shape[0]), flags=cv2.INTER_NEAREST)


    
    
    Shrinking_video.write(shrunk_image)

    



rotation_video.release()
scalling_video.release()
translation_video.release()
vertical_shear_video.release()
Horizontal_shear_video.release()
zoom_video.release()
Shrinking_video.release()


#Identity = cv2.warpAffine(image,matrix_Identity,(image.shape[1],image.shape[0]))
#Scalling = cv2.warpAffine(image,matrix_scalling,(image.shape[1]+500,image.shape[0]+500))
#rotation = cv2.warpAffine(image,matrix_rotation,(image.shape[1],image.shape[0]))
#translation = cv2.warpAffine(image,matrix_translation,(image.shape[1],image.shape[0]))



#cv2.imshow('Identity',Identity)
#cv2.imshow('Scalling',Scalling)
#cv2.imshow('rotation',rotation)
#cv2.imshow('translation',translation)
#cv2.imshow('vertical_shear',vertical_shear)
#cv2.imshow('Horizontal_shear',Horizontal_shear)
#cv2.imshow('Zooming',image_re_linear)
#cv2.imshow('shrunken',shrunk_image)
#cv2.imshow('Original',image)


# Load the individual video files
#rotation_video2 = cv2.VideoCapture('rotation_video.avi')
#vertical_shear_video2 = cv2.VideoCapture('vertical_shear_video.avi')
#Horizontal_shear_video2 = cv2.VideoCapture('Horizontal_shear_video.avi')
#scaling_video2 = cv2.VideoCapture('scalling_video.avi')
#translation_video2 = cv2.VideoCapture('translation_video.avi')
#zoom_video2 = cv2.VideoCapture('zoom_video.avi')
#Shrinking_video2 = cv2.VideoCapture('Shrinking_video.avi')




```

```python
import cv2
import numpy as np



# Load the individual video files
v1 = cv2.VideoCapture('rotation_video.avi')
v2 = cv2.VideoCapture('vertical_shear_video.avi')#cv2.VideoCapture('Horizontal_shear_video.avi')#cv2.VideoCapture('vertical_shear_video.avi')
#Horizontal_shear_video2 = cv2.VideoCapture('Horizontal_shear_video.avi')
v3 = cv2.VideoCapture('scalling_video.avi')
v4 = cv2.VideoCapture('translation_video.avi')
#zoom_video2 = cv2.VideoCapture('zoom_video.avi')
#Shrinking_video2 = cv2.VideoCapture('Shrinking_video.avi')


# Check if the videos opened successfully
if not (v1.isOpened() and v2.isOpened() and v3.isOpened() and v4.isOpened()):
    print("One or more videos could not be opened.")
    exit()

# Get the video frame dimensions
frame_width = int(v1.get(3))
frame_height = int(v1.get(4))


# Create a video writer to save the output video
output_file = "output_video_final 2.avi" # avi mp4
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # codec XVID mp4v
out = cv2.VideoWriter(output_file, fourcc, frame_rate, ( 2 * frame_width,2 * frame_height))

while True:
    ret1, frame1 = v1.read()
    ret2, frame2 = v2.read()
    ret3, frame3 = v3.read()
    ret4, frame4 = v4.read()

    if not (ret1 and ret2 and ret3 and ret4):
        break

    # Resize frames to the same dimensions
    frame1 = cv2.resize(frame1, (frame_width, frame_height))
    frame2 = cv2.resize(frame2, (frame_width, frame_height))
    frame3 = cv2.resize(frame3, (frame_width, frame_height))
    frame4 = cv2.resize(frame4, (frame_width, frame_height))
    

    # Create an empty frame with subplots
    empty_frame = np.zeros((2 * frame_height, 2 * frame_width, 3), dtype=np.uint8)

    # Place each video frame in a subplot
    empty_frame[:frame_height, :frame_width] = frame1
    empty_frame[:frame_height, frame_width:] = frame2
    empty_frame[frame_height:, :frame_width] = frame3
    empty_frame[frame_height:, frame_width:] = frame4
   
    # Write the combined frame to the output video
    out.write(empty_frame)

# Release the video capture and writer objects
v1.release()
v2.release()
v3.release()
v4.release()
out.release()

print("Output video saved as:", output_file)

if(cv2.waitKey()==ord('q')):
    cv2.destroyAllWindows()
```
